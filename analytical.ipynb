```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
t=1
a = 0.7e-9
```

```python
def spec(kx, ky, t=1):
    '''
    This is for plotting the energy spectrum in momentum space
    kx, ky are in units of a
    '''
    E1 = 0*kx
    E2 = 2*t*np.sqrt(1+1/2*(np.cos(kx)+np.cos(ky)))
    E3 = -E2
    return E1, E2, E3
```

```python
k = np.linspace(0, 2*np.pi, 1000)
k0 = np.zeros(1000)
k1 = np.ones(1000)
#hello
```

```python
plt.xlabel("$k_x$a", size='x-large')
plt.ylabel("E", size='x-large')
plt.yticks(size='x-large')
plt.xticks(ticks= [-np.pi, 0, np.pi, 2*np.pi], labels=[r"-$\pi$","0",r"$\pi$", r"2$\pi$"], size='x-large')

plt.plot(k, spec(k, k*0)[0])
plt.plot(k, spec(k, k*0)[1])
plt.plot(k, spec(k, k*0)[2])
#test
plt.savefig("Ekx.png")
plt.show()
```

```python
plt.xlabel("$k_y$a", size='x-large')
plt.ylabel("E", size='x-large')
plt.yticks(size='x-large')
plt.xticks(ticks= [-np.pi, 0, np.pi, 2*np.pi], labels=[r"-$\pi$","0",r"$\pi$", r"2$\pi$"], size='x-large')

plt.plot(k, spec(k*0, k)[0])
plt.plot(k, spec(k*0, k)[1])
plt.plot(k, spec(k*0, k)[2])

plt.savefig("Eky.png")
plt.show()
```

```python
plt.xlabel("$(k_x a,k_y a)$", size='x-large')
plt.ylabel("E", size='x-large')
plt.yticks(size='x-large')
plt.xticks(ticks= [-np.pi, 0, np.pi*np.sqrt(2), 2*np.pi*np.sqrt(2)], labels=[r"(-$\pi$,-$\pi$)","(0,0)",r"($\pi$,$\pi$)", r"(2$\pi$,2$\pi$)"], size='x-large')

plt.plot(np.sqrt(k**2*2), spec(k, k)[0])
plt.plot(np.sqrt(k**2*2), spec(k, k)[1])
plt.plot(np.sqrt(k**2*2), spec(k, k)[2])

plt.show("Ediag.png")
plt.show()
```
